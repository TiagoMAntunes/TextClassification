```python
with open('../data_0.2_256_1_glove', 'rb') as f:
    mdata = pickle.load(f)
```

```python
accs = pd.DataFrame(np.array([mdata[1], mdata[3]]).T)
```

```python
loss = pd.DataFrame(np.array([mdata[0], mdata[2]]).T)
```

```python
accs = accs.rename(columns = {0:'Train', 1:'Dev'}) * 100
```

```python
loss = loss.rename(columns = {0:'Train', 1:'Dev'})
```

```python
ax = accs.plot(xlabel="# Epochs", ylabel="Accuracy (%)")
fig = ax.get_figure()
fig.savefig('accuracyplot.png')
```

```python
ax = loss.plot(xlabel="# Epochs", ylabel="Loss")
fig = ax.get_figure()
fig.savefig('lossplot.png')
```
